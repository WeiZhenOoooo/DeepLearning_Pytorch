import package

```python
from autogluon.tabular import TabularDataset, TabularPredictor
import numpy as np
import pandas as pd
```

# 加载数据

```python
train_data = TabularDataset('/kaggle/input/california-house-prices-1/train.csv')
test_data = TabularDataset('/kaggle/input/california-house-prices-1/test.csv')
```

查看训练数据

```python
train_data.head()
```

```python
train_data.shape
```

# 处理数据

```python
id, label = 'Id', 'Sold Price'
large_vel_cols = ['Lot', 'Total interior livable area', 'Tax assessed value',
                  'Annual tax amount', 'Listed Price', 'Last Sold Price']
for c in large_vel_cols:
    train_data[c] = np.log(train_data[c] + 1)
```

# 训练

```python
predictor = TabularPredictor(label=label).fit(train_data.drop(columns=[id]))
```

# 预测

```python
preds = predictor.predict(test_data.drop(coulums=[id]))
submission = pd.DataFrame({id:test_data[id], label:preds})
submission.tocsv('submission.csv', index=False)
```
